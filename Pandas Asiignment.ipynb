```python
#ques12
def select_rows_between_dates(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Select rows between '2023-01-01' and '2023-01-31'
    selected_rows = df[(df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')]
    
    return selected_rows

# Example usage:
# Assuming you already have a DataFrame named 'df' with a 'Date' column
# Call the function to select rows between '2023-01-01' and '2023-01-31'
selected_df = select_rows_between_dates(df)
```
