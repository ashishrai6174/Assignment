```python
#ques1:
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Simple Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a scatter glyph
source = ColumnDataSource(data=dict(x=x, y=y))
p.scatter(x='x', y='y', source=source, size=10, color="navy")

# Show the plot in a Jupyter Notebook (or use "output_file" to save as an HTML file)
output_notebook()
show(p)
```

```python
#ques 2:
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Scatter Plot with Glyphs", x_axis_label='X-axis', y_axis_label='Y-axis')

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y))

# Add a circle glyph to the figure
p.circle('x', 'y', source=source, size=10, color="navy", alpha=0.6)

# Show the plot in a Jupyter Notebook (or use "output_file" to save as an HTML file)
output_notebook()
show(p)
```

```python
#ques 3:
from bokeh.plotting import figure, show

# Create a Bokeh figure
p = figure(title="Customizing Axes", x_axis_label='X-axis', y_axis_label='Y-axis')

# Customize x-axis
p.xaxis.axis_label_text_color = "red"
p.xaxis.major_label_text_font_size = "14pt"
p.xaxis.major_label_orientation = "horizontal"
p.xaxis.major_tick_line_color = "green"
p.xaxis.major_tick_out = 10

# Customize y-axis
p.yaxis.axis_label_text_color = "blue"
p.yaxis.major_label_text_font_size = "14pt"
p.yaxis.major_label_orientation = "vertical"
p.yaxis.major_tick_line_color = "purple"
p.yaxis.major_tick_in = 5

# Add data and glyphs here

# Show the plot
show(p)
```
